```python
import pandas as pd

#1. Criando os dados (dicionario)
dados = {
  'Vendedor': ['Ana', 'Bia', 'Caio', 'Duda', 'Enzo'],
  'Vendas': [150, 400, 50, 250, 600] 
}

#2. Transformando em um DataFrame (Tabela de Pandas)
df =pd.DataFrame(dados)

#3. Fasendo calculos Rápidos
media_vendas = df['Vendas'].mean()
total_vendas = df['Vendas'].sum()

#4. Exibindo a tabela e os resultados
print(f"Média de Vendas: R$ {media_vendas}")
print("_" * 20)
print(df)
print(f"Vendas acima de R$ 300")
print(df[df['Vendas'] > 300])

import matplotlib.pyplot as plt 
#1. Definir o que vai no eixo x (Vendedor) e no eixo y (Vendas)
plt.bar(df['Vendedor'], df['Vendas'], color='skyblue')

#2. Asicinar títulos para ficar profissinal
plt.title('Desempenho de Vendas por Vendedor')
plt.xlabel('Vendedor')
plt.xlabel('Vendas')

#3. Mostrar o gráfico
plt.show()
```
