```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

#1. Criando nosso Banco de Dado de treino  
# Features(caractéristicas): Meses de contrato, Valor mensal, Reclamação abertas 
# Target (Alvo): Cancelou(1 para Sim, 0 para Não)
dados_churn = {
    'Meses_Contrato': [1, 12, 2, 24, 3, 36],
    'Valor_Mensal': [100, 80, 110, 70, 120, 60],
    'Reclamacoes': [3, 0, 2, 0, 4, 1],
    'Cancelou': [1, 0, 1, 0, 1, 0]
    }

df_churn = pd.DataFrame(dados_churn)

# 2. Treinando a IA
X = df_churn[['Meses_Contrato', 'Valor_Mensal', 'Reclamacoes']]
y = df_churn['Cancelou']

modelo_churn = RandomForestClassifier(random_state=42)
modelo_churn.fit(X, y)

print("IA treinada para prever cancelamentos!")
print("_" * 30)

#3. O teste real: Previsão para um Novo Cliente
#Cliente novo: 1 mês de contrato, paga R$ 115,00, e já fez 3 reclamações.
novo_cliente = [[1, 115, 3]]
previsao = modelo_churn.predict(novo_cliente)

if previsao[0] == 1:
 print("ALERTA: Este cliente tem ALTA chance de cancelar!")
 print("Sugestão: Ofereça um desconto urgente.")
else:
 print("Cliente estável: Baixa chance de cancelamento.")
```
