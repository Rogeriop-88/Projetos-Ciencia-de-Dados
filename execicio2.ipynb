```python
#1.Crie a lista de vendas
vendas = [150, 400, 50, 250, 600]

#2.calcule o total e a maior venda
total = sum(vendas)
maior_valor = max(vendas)

#3. Exibir os resultados
print(f"O total de vendas foi: R${total}")
print(f"A maior venda registrada foi: R${maior_valor}")

#4. Verifique a meta(complete o cÃ³dico abaixo)
if total > 1000:
    print("Meta batida!")
else:
    print("Abaixo da meta")    
```
